```python
from google.cloud import aiplatform
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_google_vertexai import VertexAIEmbeddings
from langchain.vectorstores import FAISS

# Initialize Google Vertex AI
aiplatform.init(project="composed-maxim-443209-r4", location="us-central1")

# Load and Split Text
text = """LangChain allows developers to connect LLMs with data sources and APIs.
It supports features such as embeddings and vector databases."""
splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=10)
docs = splitter.create_documents([text])

# Generate Embeddings
embedding = VertexAIEmbeddings(model_name="text-embedding-004")
vectors = [embedding.embed_query(doc.page_content) for doc in docs]

# Store in Vector Database (e.g., FAISS)
vector_db = FAISS.from_documents(docs, embedding)
query_result = vector_db.similarity_search("What is LangChain?", k=1)

# Output Results
print("Documents:", docs)
print("Vectors:", vectors)
print("Search Result:", query_result)
```
